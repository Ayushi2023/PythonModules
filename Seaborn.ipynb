* SWARM PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style = "whitegrid")
iris = sns.load_dataset("iris")
#print(iris)

ax = sns.swarmplot(x = "species", y = "sepal_width", data = iris)
plt.title("Graph")
plt.show()
```

* COUNT PLOT

```python
df = sns.load_dataset("iris")   ## reading the datasets

sns.countplot(x = "species", data = df)
plt.show()
```

```python
df = sns.load_dataset("tips")

sns.countplot(x = "sex", data = df)
plt.show()
```

* BOX PLOT

```python
df = sns.load_dataset("tips")   ## loading dataset
sns.boxplot(x = "day", y = "total_bill",data = df,hue = "smoker")       
plt.show()
                ## points outside the box(represents it is out from the range)
```

* VIOLIN PLOT

```python
df = sns.load_dataset("tips")
sns.violinplot(x = "day", y = "total_bill",data = df, hue = "sex", split = True )
plt.show()
```

* STRIP PLOT

```python
df = sns.load_dataset("tips")   ## loading dataset
sns.stripplot(x = "day", y = "total_bill",data = df,jitter = True, hue = "smoker", dodge = True)
plt.show()
```
