## 1. Create array linearly spaced 20 elements from 10 to 25 and plot their graph against elements from 1 to 20.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(10, 25, 0.75)
y = np.array([i for i in range(20)])

plt.plot(x, y)
plt.show()
```

## 2. Write a program to draw a sine wave for any 10 values.

```python
x = np.arange(0, np.pi * 5, 0.25)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

## 3.Draw the subplots for x vs. y and x vs. t for the following values:
* x = [12,34,54,34,67,68,59,89,92,44] 
* y = [11,19,26,37,58,69,88,103,111,128]
* t = [2,4,8,12,22,44,68,70,100,120]

```python
x = np.array([12,34,54,34,67,68,59,89,92,44])
y = np.array([11,19,26,37,58,69,88,103,111,128])
t = np.array([2,4,8,12,22,44,68,70,100,120])

plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,t)
plt.show()
```

## 4. Following is the sales information in lacs rupees from year 2010 to 2017 for an e-commerce store. Draw the bar chart for same.
* 123, 214,233,255,289,273,167,304

```python
y = np.array([123, 214,233,255,289,273,167,304])
x = np.array([i for i in range(2010, 2018)])

plt.bar(x, y)
plt.ylabel("Rupees in Lacs")
plt.xlabel("Years")
plt.title("Sales Information", color = "red")
plt.show()
```
